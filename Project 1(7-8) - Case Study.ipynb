```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
startup = pd.read_csv("startup_funding.csv", encoding='latin1')
df = startup.copy()
df.loc[df["StartupName"].isin(["Ola Cabs","Olacabs"]),"StartupName"]="Ola"
df.loc[df["StartupName"].isin(["Flipkart.com"]),"StartupName"]="Flipkart"
df.loc[df["StartupName"].isin(["Oyo Rooms","OYOfit","Oyorooms","OyoRooms"]),"StartupName"]="Oyo"
df.loc[df["StartupName"].isin(["Paytm Marketplace","Paytm"]),"StartupName"]="Paytm"
Fundings=df.groupby("StartupName")["Date"].count()
company = Fundings.index
count_fund = Fundings.values
np_company = np.array(company)
np_count_fund = np.array(count_fund,dtype="int64")
np_company = np_company[np.argsort(np_count_fund)][:-6:-1]
np_count_fund = np.sort(np_count_fund)[:-6:-1]
for i in range(len(np_company)):
    print(np_company[i],np_count_fund[i])
plt.bar(np_company,np_count_fund,width = 0.5)
plt.xticks(rotation=90)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
startup = pd.read_csv("startup_funding.csv", encoding='latin1')
df = startup.copy()
df.dropna(subset=["InvestorsName"],inplace = True)
np_investors = np.array(df["InvestorsName"],dtype='str')
investors =[]
for row in np_investors:
    if "undisclosed" not in row and "Undisclosed" not in row:
        val = row.split(",")
        for name in val:
            investors.append(name.strip())
dict={}
for i in investors:
    if i  in dict:
        dict[i]+=1
    else:
        dict[i]=1
x=[]
y=[]
for key,values in dict.items():
    x.append(key)
    y.append(values)
np_x = np.array(x)
np_y = np.array(y)
np_x = np_x[np.argsort(np_y)][:-2:-1]
np_y = np.sort(np_y)[:-2:-1]
print(np_x[0],int(np_y[0]))
```
